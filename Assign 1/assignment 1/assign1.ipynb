### KNN implementation

###  Q 1. a)  Validation set Performance of 7 nearest neighbour regression model using RMSE

$$ RMSE=\sqrt{\sum_{i=1}^{N} \frac{{(Ypredicted -Yactual)}^2}{N} }$$
$$ RMSE=Root Mean Square Error$$

```python
def knn(Xtrain,Xtest,Ytrain,Ytest,k):
    Ypred = [] # list to store prediction for all test points
    # iterating over all test input
    for i in range(len(Xtest)):
        distance = np.sqrt(np.sum((Xtrain - Xtest[i])**2 ,axis=1))  # calculating Euclidean distance for test input from all points in train data
        index=np.argsort(distance) # use argsort to get indices of sort distances
        k_index=index[:k] #select k least distances
        avg=Ytrain[k_index].mean() #Selecting K nearest point and finding averge of their outputs.
        Ypred.append(avg)  #append predicted price into Ypred list
    n=len(Ypred) # getting length of Ypred 
    rmse=np.sqrt(np.sum((Ytest - Ypred)**2)/ n) #Find root mean square error between predicted and actual output
    return rmse 
```

Use above KNN method to Find Root mean square error 

```python
print("RMSE for K=7 is" ,knn(Xtrain,Xtest,Ytrain,Ytest,7))
```

### Q 1. b) Optimal values  of K is 3 using Cross validation

```python
kvalues = [x for x in range(1,15,1)] # creating list containing values from 1 to 14
rmselist=[]   #list to store RMSE values

for i in kvalues:
    rmselist.append(knn(Xtrain,Xtest,Ytrain,Ytest,i))
                 
index=np.argsort(rmselist) #sorting rlist to find least RSME values 
idx=index[0] #getting index of least Root Mean Square error
print("Least RMSE Values is",rmselist[idx])
print("Best K is ",kvalues[idx])
```

```python
plot.figure(figsize=(10,6))
plot.plot(kvalues, rmselist,'r')
plot.scatter(kvalues[idx],rmselist[idx])
plot.ylim(2000,5000)
Point=[kvalues[idx],rmselist[idx]]
plot.annotate(Point, (kvalues[idx],rmselist[idx]))
plot.xlabel('K value')
plot.ylabel('RMSE')
plot.title('RMSE vs K in KNN')
plot.show()
```

#### Above graph is between K values and RMSE with point  at which K is optimal i.e K=3

```python
class Node:
    #Node store depth of tree , stump , parent , child nodes
    #leaf node store mojarity of Votes
    def __init__( self, depth = 0, stump = (0,0), parent = None ):
        self.depth = depth
        self.stump = stump
        self.parent = parent
        self.left = None
        self.right = None
        self.isLeaf = True
        self.label = 0
        
        
    def predict( self, data ):
        if self.isLeaf: # if we at leaf then return predicted value
            return self.label
        else: #If not then goes into right / left node according stump values
            if data[self.stump[0]] > self.stump[1]:
                return self.right.predict( data )
            else:
                return self.left.predict( data )
            
     #calculate Entropy of perticular features values       
    def entropy(self,s):
        cnt_uniq = np.unique(s,return_counts=True) #count unique feature values  and their count
        N = int(s.shape[0])
        entr = 0.0
        for ix in cnt_uniq[1]: #iterate over all unique values
            frac  = ix/N   
            entr +=  frac * np.log2(frac)
        entr=-entr
        return entr  #returning entropy
    
    #calculating Information gain for given Feature 
    def information_gain(self, parent, left_child, right_child):
        leftf = len(left_child) / len(parent) # count length of left child feature
        rightf = len(right_child) / len(parent) # count length of right child feature
        left_entropy=self.entropy(left_child) #calculating left child entropy
        right_entropy=self.entropy(right_child)#calculating right child entropy
        infogain=self.entropy(parent)-(leftf*left_entropy + rightf*right_entropy) #calculating information gain 
        return infogain
    
    #spliting values according to theresold values i.e feature value
    def split(self,dataset,feature,feature_value):
        li=[]
        li1=[]
        for i in dataset:
            if i[feature]<=feature_value: #if value <= then add it into left otherwise into left 
                li.append(i)
            else:
                li1.append(i)
        return (np.array(li),np.array(li1))
    
    # Only one feature is tested at each node of decision tree 
    # getstump use to find feature which having most Info_gain and return theresold values
    def getStump( self, X, y ):
        best_gain = -1
        best_feature = 0
        for feature in range(X.shape[1]): #iterating over all features
            X_curr=X[:,feature]
            for fea_value in np.unique(X_curr): #selecting unique values from each feature
                y1=y
                df = np.concatenate((X, y1.reshape(1, -1).T), axis=1) #finding corresponding target values
                df_left,df_right=self.split(df,feature,fea_value) 
                if len(df_left) > 0 and len(df_right) > 0:  
                    y = df[:, -1]
                    y_left = df_left[:, -1]
                    y_right = df_right[:, -1]
                    gain = self.information_gain(y, y_left, y_right)
                    if gain > best_gain:# getting best gain
                        best_gain=gain
                        best_feature=feature
                        threshold=fea_value
        return (best_feature,threshold) #returning best feature anf thersolf value

    def train( self, X, y, maxLeafSize, maxDepth ):
        if y.size < maxLeafSize or self.depth >= maxDepth:
            self.isLeaf = True
            self.label = np.bincount(y).argmax() #finding majority votes using numpy
        else:
            self.isLeaf = False
            self.stump = self.getStump( X, y ) #getting best feature and thersold values for that feature
            self.left = Node( depth = self.depth + 1, parent = self ) 
            self.right = Node( depth = self.depth + 1, parent = self )
            discriminant = X[:, self.stump[0]] - self.stump[1] 
            # if values is less than thresold the point goes to left othervise
            self.left.train( X[discriminant <= 0, :], y[discriminant <= 0], maxLeafSize, maxDepth )
            self.right.train( X[discriminant > 0, :], y[discriminant > 0], maxLeafSize, maxDepth )

class Tree:
    def __init__( self, maxLeafSize , maxDepth  ):# define contructor forgetting maxleafsize and maxDepth
        self.root = Node() 
        self.maxLeafSize = maxLeafSize
        self.maxDepth = maxDepth 
     #for predicting the output for given input data   
    def predict( self, X ):
        return self.root.predict(X) 
    #training model using training data
    def train( self, X, y ):
        self.root.train( X, y, self.maxLeafSize, self.maxDepth )
```
